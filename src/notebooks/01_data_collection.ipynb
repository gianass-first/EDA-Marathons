```python
import numpy as np

# -----------------------
# DATOS TOKYO (MEN)
# -----------------------

men = np.array([
    1644,                         # <3h
    7684 - 1644,                  # 3–4h
    8028 - 7684,                  # 4–5h
    5927,                         # 5–6h
    3424                          # 6–7h
])

# -----------------------
# DATOS TOKYO (WOMEN)
# -----------------------

women = np.array([
    133,                          # <3h
    1658 - 133,                   # 3–4h
    3139 - 1658,                  # 4–5h
    2777,                         # 5–6h
    1723                          # 6–7h
])

# -----------------------
# TIEMPOS MEDIOS POR RANGO
# -----------------------

tiempos = np.array([2.5, 3.5, 4.5, 5.5, 6.5])  # horas

# -----------------------
# MEDIA PONDERADA
# -----------------------

average_men_time = np.sum(men * tiempos) / np.sum(men)
average_women_time = np.sum(women * tiempos) / np.sum(women)

print("Average men time (hours):", average_men_time)
print("Average women time (hours):", average_women_time)
```

```python
def hours_to_hhmm(h):
    hours = int(h)
    minutes = int((h - hours) * 60)
    return f"{hours:02d}:{minutes:02d}"

print("Men:", hours_to_hhmm(average_men_time))
print("Women:", hours_to_hhmm(average_women_time))
```
